```python
import matplotlib.pyplot as plt

lines = open("./numberK.txt", "r").readlines()
K = []
Sum_of_squared_distances = []

for line in lines:
    line = line.replace("\n", "")
    tmp = line.split(",")
    K.append(int(tmp[0]))
    Sum_of_squared_distances.append(float(tmp[1]))

plt.figure(figsize = (11,11))
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

lines = open("./partialExample/part-00000", "r").readlines()

v = []

for line in lines:
    array_tmp = []
    tmp = line.split(",")
    array_tmp.append(int(tmp[0]))
    array_tmp.append(float(tmp[1]))
    array_tmp.append(float(tmp[2]))
    array_tmp.append(float(tmp[3]))
    v.append(array_tmp)

df = pd.DataFrame(v, columns=['Cluster', 'Feature1','Feature2',"Feature3"])
print (df)

fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
x = np.array(df['Feature1'])
y = np.array(df['Feature2'])
z = np.array(df['Feature3'])

ax.scatter(x,y,z, marker="s", c=df["Cluster"], s=40)

plt.show()
```
